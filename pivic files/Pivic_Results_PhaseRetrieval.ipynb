## Reprodução dos Resultados Realizados durante o PIVIC

- Neste Notebook é possível reproduzir os resultados desenvolvidos no trabalho do PIVIC

```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')

import matplotlib as mpl
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.size'] = 10
plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['text.usetex'] = False

figurePath = r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Figuras\Pivic'
```

```python
def loadArray(file_path):
    with open(file_path, 'rb') as stream:
        data = np.load(stream)
    return data
```

## Gera superfície da rede neural para as análises de BER vs FO vs Roll-off

```python
N = 256  # number of input amplitude samples to the NN

# calculates the average of BER arrays
BER_NN_01 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_NN_Ind0_SpS_4_256_Sample.npy')
BER_NN_02 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_NN_Ind1_SpS_4_256_Sample.npy')
BER_NN_03 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_NN_Ind2_SpS_4_256_Sample.npy')
BER_NN_04 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_NN_Ind3_SpS_4_256_Sample.npy')
BER_NN_05 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_NN_Ind4_SpS_4_256_Sample.npy')

BER_NN = np.mean([BER_NN_01, BER_NN_02, BER_NN_03, BER_NN_04, BER_NN_05], axis=0)

x = np.arange(0.01, 0.99, 0.1)
y = np.arange(1e9, 6e9, 1e9)

x, y = np.meshgrid(x, y)
z = np.log10(BER_NN)  # shape((5, 10))

fig = plt.figure(figsize=(14, 8))
ax =  plt.axes(projection='3d')

fig = plt.figure(figsize=(5, 5))
plt.contourf(x, y, z, cmap=cm.viridis)
plt.xlabel('Roll-off')
plt.ylabel('FO')
plt.savefig(r'C:/Users/Silas/Documents/PIVIC-Comunicacoes-Opticas/resultados/Figuras/results_ModelPerformanceContourf_64QAM.png', dpi=600)

surf = ax.plot_surface(x, y, z, cmap=cm.viridis, linewidth=0.5, antialiased=True, edgecolor='None')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=7)
ax.set_xlabel('roll-off')
ax.set_ylabel('FO')
ax.set_zlabel(r'$log(BER)$')
ax.set_title('Superfície Gerada Para BER - 64QAM')
ax.view_init(elev=30, azim=215)

plt.savefig(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Figuras\Pivic\results_ModelPerformance_64QAM_Mean.png', dpi=300)
```

## Gera superfície do algoritmo de KK para as análises de BER vs FO vs Roll-off

```python
# calculates the average of BER arrays
BER_KK_01 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_KK_Ind0_SpS_4.npy')
BER_KK_02 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_KK_Ind1_SpS_4.npy')
BER_KK_03 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_KK_Ind2_SpS_4.npy')
BER_KK_04 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_KK_Ind3_SpS_4.npy')
BER_KK_05 = loadArray(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Data\Pivic\BER_KK_Ind4_SpS_4.npy')

BER_KK = np.mean([BER_KK_01, BER_KK_02, BER_KK_03, BER_KK_04, BER_KK_05], axis=0)

x = np.arange(0.01, 0.99, 0.1)
y = np.arange(1e9, 6e9, 1e9)

x, y = np.meshgrid(x, y)
z = np.log10(BER_KK)  # shape((5, 10))

fig = plt.figure(figsize=(14, 8))
ax =  plt.axes(projection='3d')

fig = plt.figure(figsize=(5, 5))
plt.contourf(x, y, z, cmap=cm.viridis)
plt.xlabel('roll-off')
plt.ylabel('FO')
plt.savefig(r'C:/Users/Silas/Documents/PIVIC-Comunicacoes-Opticas/resultados/Figuras/results_ModelPerformanceContourf_64QAM.png', dpi=600)

surf = ax.plot_surface(x, y, z, cmap=cm.viridis, linewidth=0.5, antialiased=True, edgecolor='None')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=7)
ax.set_xlabel('roll-off')
ax.set_ylabel('FO')
ax.set_zlabel(r'$log(BER)$')
ax.set_title('Superfície Gerada Para BER - 64QAM')
ax.view_init(elev=30, azim=215)

plt.savefig(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Figuras\Pivic\results_KKPerformance_64QAM_Mean.png', dpi=600)
```

## Gera a sobreposição das superfícies entre os dois algoritmos 

```python
x = np.arange(0.01, 0.99, 0.1)
y = np.arange(1e9, 6e9, 1e9)

x, y = np.meshgrid(x, y)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, np.log10(BER_NN), alpha = 1, rstride=1, cstride=1, cmap=cm.winter, linewidth=0.5, antialiased=True, zorder = 0.5)
ax.plot_surface(x, y, np.log10(BER_KK), alpha = 0.95, rstride=1, cstride=1, cmap=cm.autumn, linewidth=0.5, antialiased=True, zorder = 0.3)
ax.set_title('Sobreposição Das Superfícies')
ax.set_xlabel('roll-off')
ax.set_ylabel('FO')
ax.set_zlabel('$log(BER)$')
ax.view_init(elev=25, azim=125)

# fake legends
fake2Dline = mpl.lines.Line2D([0],[0], linestyle="none", c='b', marker = 'o')
fake2Dline2 = mpl.lines.Line2D([0],[0], linestyle="none", c='r', marker = 'o')
ax.legend([fake2Dline, fake2Dline2], ['Rede Neural', 'Algoritmo de KK'], numpoints = 1)

plt.savefig(r'C:\Users\Silas\Documents\PIVIC-Comunicacoes-Opticas\resultados\Figuras\Pivic\results_SuperpositionSurfaces_64QAM.png', dpi=600)
```
